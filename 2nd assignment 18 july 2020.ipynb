# Random walk in 2D

In mathematics, a random walk is a mathematical object, known as a stochastic or random process, that describes a path that consists of a succession of random steps on some mathematical space such as the integers.

If we take simple case, one dimensional random walk, imagine we are  standing at the point 0 on a number line or balance beam, where you have two possible directions for movement i.e,forwards or backwards. Flip a coin to choose your direction, heads is forward, tails is back. After every flip of the coin, you take just one step; then flip again.

To imagine a random walk in two dimensions, imagine you had four possible directions you could move â€” right, left, forward or back. To choose the direction for each step, you flip a coin twice. The the first flip decides between right/left and forward/back, the next one decides which of the two directions you go.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import random as random
import numpy as np
import pandas as pd
%matplotlib inline
sns.set()
```

```python
logfile = open("data/randwalk_log.txt",'w')

'''if you are in (xi,yi) position go to (xf,yf)'''

'''set up xi,yi'''
x = 0
y = 0

'''store x and y into X and Y list'''
X = [x]
Y = [y]

''' iterate to N steps using for loop'''

for t in range(100000):
    
    '''generate a random variable'''
    r = random.uniform(0,50)
    direction =  random.choice(['x','y'])
    if direction == 'x':
            r = random.choice([-1,1])
            x = X[t-1] +r
            y = Y[t-1]
    if direction == 'y':
            r = random.choice([-1,1])
            y = Y[t-1] + r
            x = X[t-1]
    
    '''remember to update X and Y in every loop'''
    X.append(x)
    Y.append(y)
    
logfile.close()
```

```python
Data = {"x":X,"y":Y}
import json as json
with open("data/randwalk_dictionary.json",'w') as f:    # saving random walk data in json file
    json.dump(Data,f)
```

```python
DF = pd.DataFrame(Data)
DF.head()
```

```python
DF.to_csv('data/randwalk_dataframe.csv')  # saving random walk data in csv file
```

```python
plt.figure(figsize = [8,8])
plt.scatter(X,Y,color = 'b',marker = '.')
plt.savefig("plot/randwalk.pdf")      #saving plot of random walk data in pdf
plt.savefig("plot/randwalk.png")      #saving plot of random walk data in png
plt.show()
```
